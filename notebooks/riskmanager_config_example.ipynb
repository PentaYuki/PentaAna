# Cấu hình RiskManager và LiveTradingEngine
Ví dụ cấu hình các mức rủi ro khác nhau cho Stock-AI.

```python
from live_broker import RiskManager, LiveTradingEngine, PaperBroker
# Cấu hình thận trọng: risk 1% / lệnh, max 30tr
conservative = RiskManager(
    account_value_vnd  = 500_000_000,
    risk_per_trade_pct = 1.0,       # 1% vốn / lệnh (mặc định 1.5%)
    max_position_vnd   = 30_000_000,
    max_daily_orders   = 3,         # Tối đa 3 lệnh/ngày
    atr_stop_mult      = 1.2,       # Stop gần hơn
    atr_trail_mult     = 1.8,
    max_slippage_pct   = 0.3,       # Slippage chặt hơn
)
# Cấu hình tích cực: risk 2% / lệnh, max 80tr
aggressive = RiskManager(
    account_value_vnd  = 500_000_000,
    risk_per_trade_pct = 2.0,
    max_position_vnd   = 80_000_000,
    max_daily_orders   = 8,
    atr_stop_mult      = 2.0,       # Stop rộng hơn
    atr_trail_mult     = 3.0,
)
# Khởi tạo engine với cấu hình mong muốn
engine = LiveTradingEngine(
    broker       = PaperBroker(),
    risk_manager = conservative,   # Đổi sang aggressive khi muốn
)
```
